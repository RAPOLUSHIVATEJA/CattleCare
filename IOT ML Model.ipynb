```python
# Seperating the independent features
X = df.iloc[ : , :-1].values
print(X)
```

```python
# Seperating the dependent feature
y = df.iloc[ : , -1].values
print(y)
```

```python
# Mean encoding to fill the NAN values
from sklearn.impute import SimpleImputer  # We are importing SimpleImputer class from sklearn.impute module.
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(X[:, :2])  # Applying this imputer object on the matrix of features
imputer.transform(X[:, :2]) # Calling the transformer object so that the missing values will be replaced by the mean of the respective feature
X[:, :2] = imputer.transform(X[:, :2])
```

```python
print(X)
```

```python
# Encoding the dependent variable
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
print(y)
```

# Splitting the data into training test and test set

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split( X , y , test_size = 0.25 , random_state = 1)
```

```python
print(X_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

# Feature scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler() # We should not apply feature scaling on the dummy variables
```

```python
X_train[ : , :3 ] = sc.fit_transform(X_train[ : , :3 ])
```

```python
X_test[ : , :3 ] = sc.fit_transform(X_test[ : , :3 ])
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train,y_train)
```

```python
y_predicted_1 = lr.predict(X_test)
```

```python
y_predicted_1
```

```python
accuracy_score_1 = lr.score(X_test , y_test)
accuracy_score_1
```

```python
from sklearn.metrics import f1_score
f1_score(y_test,y_predicted_1,average=None)
```

```python
from sklearn import metrics
cm = metrics.confusion_matrix(y_test , y_predicted_1) 
cm
```

```python
plt.figure(figsize=(9,9))
sns.heatmap(cm, annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r')
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
plt.title('Confusion Matrix')
```

# Support Vector Classification

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(X_train, y_train)
```

```python
y_predicted = svc.predict(X_test)
```

```python
y_predicted
```

```python
score = svc.score(X_test , y_test)
score
```

```python
from sklearn.metrics import accuracy_score
accuracy_score_2 = accuracy_score(y_test, y_predicted)
accuracy_score_2
```

# Making the Confusion matrix

```python
from sklearn import metrics
cm = metrics.confusion_matrix(y_test , y_predicted) 
cm
```

```python
plt.figure(figsize=(9,9))
sns.heatmap(cm, annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r')
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
plt.title('Confusion Matrix')
```

# KNN Classification

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=11)
knn.fit(X_train,y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score_3 = accuracy_score(y_test, y_pred)
accuracy_score_3
```
